```python
from sklearn.datasets import load_diabetes
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# data = load_boston()
df = load_diabetes()
df
```

```python
type(df)
```

```python
dataset = pd.DataFrame(df.data)
dataset.columns = df.feature_names #adding independent variables
print(dataset.head())
```

```python
dataset["Diabetes Progression"] = df.target #adding dependent variable
dataset.head()
```

```python
X = dataset.iloc[:,:-1] #independent features
Y = dataset.iloc[:,-1] #dependent features
```

```python
X.head()
# Y.head()
```

### SKlearn Liner Regression

```python
#Cross validation helps in selection of test and train data by altering the ratio each time (eg: 80/20, 50/40 ...)

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

LinReg = LinearRegression()
mse = cross_val_score(LinReg,X,Y, scoring='neg_mean_squared_error', cv=5) 
mean_mse = np.mean(mse)
print("The cross validation values are: ",mse)
print("The mean is ", mean_mse) #negative value will be present due to neg_mean_squared_error
```

```python
LinReg.fit(X,Y)
```

```python
prediction = LinReg.predict(X)
prediction
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.30, random_state=None)
```

#### After split of data

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
LinReg = LinearRegression()
mse = cross_val_score(LinReg, X_train, y_train, scoring='neg_mean_squared_error', cv = 5)
mean_mse = np.mean(mse)
print("The cross validation values are: ",mse)
print("The mean is ", mean_mse)
```

### Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV #it will help with hyper parameter tuning
ridge = Ridge() 
params = {'alpha':[1e-15, 1e-10, 1e-8, 1e-6, 1e-4, 1e-3, 1e-2, 1, 2, 5, 10, 100]}
ridge_regressor = GridSearchCV(ridge, params, scoring='neg_mean_squared_error', cv = 5)
ridge_regressor.fit(X,Y)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

#### After split of data

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge = Ridge()
params = {'alpha' :[1e-15, 1e-10, 1e-8, 1e-6, 1e-4, 1e-3, 1e-2, 1, 2, 5, 10, 100]}
ridge_regressor = GridSearchCV(ridge, params, scoring='neg_mean_squared_error', cv = 5)
ridge_regressor.fit(X_train, y_train)

print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

### Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

lasso = Lasso()
params = {'alpha':[1e-15, 1e-10, 1e-8, 1e-6, 1e-4, 1e-3, 1e-2, 1, 2, 5, 10]}
lasso_regressor = GridSearchCV(lasso, params, scoring='neg_mean_squared_error', cv = 5)
lasso_regressor.fit(X,Y)
```

```python
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

#### After split of data

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

lasso = Lasso()
params = {'alpha':[1e-15, 1e-10, 1e-8, 1e-6, 1e-4, 1e-3, 1e-2, 1, 2, 5, 10, 100]}
lasso_regressor = GridSearchCV(lasso, params, scoring='neg_mean_squared_error', cv=5)
lasso_regressor.fit(X_train, y_train)

print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

```python
y_pred = lasso_regressor.predict(X_test)
from sklearn.metrics import r2_score

r2_score1 = r2_score(y_pred, y_test)
print(r2_score1)
```

```python
from sklearn.metrics import r2_score

def adjusted_r2_score(y_pred, y_test, n, p):
    r2 = r2_score(y_pred, y_test)
    return 1- (((1-r2)*(n-1))/n-p-1)

# Example usage:

n = len(y_test)  # Number of data points
p = 8  # Number of predictors (e.g., features)

adjusted_r2 = adjusted_r2_score(y_pred, y_test, n, p)
print("Adjusted R² Score:", adjusted_r2)


```
